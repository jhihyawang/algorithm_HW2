```python
import time
import matplotlib.pyplot as plt

def fibo(n):
    if n <= 1:
        return n
    return fibo(n-1)+fibo(n-2)

n_values = list(range(1, 101))
execution_times = []

for n in n_values:
    start_time = time.time()
    fibo(n)
    end_time = time.time()
    execution_time = end_time - start_time
    execution_times.append(execution_time)
    print(n, execution_time)

    if execution_times[-1] > 43200:
        max_n = n
        break
else:
    max_n = 100

plt.plot(n_values[:max_n], execution_times, marker='o')
plt.xlabel('n')
plt.ylabel('Execution Time (s)')
plt.title('Execution Time of Fibonacci Function')
plt.grid(True)
plt.show()

print("Maximum value of n reached:", max_n)
```

```python
#Top-down Dynamic Programming

import time
import matplotlib.pyplot as plt

def fibonacci(n):
    cache = {0: 0, 1: 1, 2: 1}

    def fib(n):
        if n in cache:
            return cache[n]
        cache[n] = fib(n - 1) + fib(n - 2)
        return cache[n]
    return fib(n)

n_values = list(range(1, 101))
execution_times = []

for n in n_values:
    start = time.time()  # Record start time for each iteration
    fibonacci(n)  # Call fibonacci function to calculate Fibonacci number
    end = time.time()  # Record end time for each iteration
    execution_time = end - start  # Calculate execution time for each iteration
    execution_times.append(execution_time)
    print(n, execution_time)

# Plot the results
plt.plot(n_values, execution_times, marker='o')
plt.xlabel('n')
plt.ylabel('Execution Time (s)')
plt.title('Execution Time of Top-down Method')
plt.grid(True)
plt.show()
```

```python
#Bottom-up Dynamic Programming
import time
import matplotlib.pyplot as plt

def fib(n):
    table = [0, 1, 1]  # dp table
    for i in range(3, n + 1):
        table.append(table[i - 1] + table[i - 2])  # fill the table
    return table[n]
n_values = list(range(1, 101))
execution_times = []
for n in n_values:
    start = time.time()  # Record start time for each iteration
    fib(n)  # Call fib function to calculate Fibonacci number
    end = time.time()  # Record end time for each iteration
    execution_time = end - start  # Calculate execution time for each iteration
    execution_times.append(execution_time)
    print(n, execution_time)
    
# Plot the results
plt.plot(n_values, execution_times, marker='o')
plt.xlabel('n')
plt.ylabel('Execution Time (s)')
plt.title('Execution Time of Bottom-up Method')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

def fib_recursive(n):
    if n <= 1:
        return n
    
    f4_count = 0
    
    def helper(n):
        nonlocal f4_count
        if n == 4:
            f4_count += 1
        if n <= 1:
            return n
        return helper(n - 1) + helper(n - 2)
    
    fib_n = helper(n)
    
    print(f" F({n})  F(4) : {f4_count}")
    
    return fib_n, f4_count

results = [fib_recursive(i)[1] for i in range(5, 51)]

plt.plot(range(5, 51), results, marker='o')
plt.xlabel('Fibonacci number (n)')
plt.ylabel('Number of times F(4) is computed')
plt.title('Number of times F(4) is computed from F(5) to F(50) (Recursive)')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

def fib_with_overlapping_subproblems(n):
    table = [0, 1, 1]  
    f4_count = 0  
    for i in range(3, n+1):
        table.append(table[i - 1] + table[i - 2]) 
        #print("F",n,"F",i+1,"=",table[i - 1],"+",table[i - 2])
        if i >= 5 and table[i - 1] == 3:  
            f4_count += 1
    #print(f" F({n})  F(4) : {f4_count}")
    return f4_count

counts = [fib_with_overlapping_subproblems(i) for i in range(5, 51)]

plt.plot(range(5, 51), counts, marker='o')
plt.xlabel('Fibonacci number (n)')
plt.ylabel('Number of times F(4) is computed')
plt.title('Number of times F(4) is computed from F(5) to F(50) (Dynamic Programming)')
plt.grid(True)
plt.show()
```
